```python
import numpy as np
import matplotlib.pyplot as plt

N = 100

mean = 0 
std_dev1 = 1
std_dev2 = 2

random_signal1 = np.random.normal(loc=mean, scale=std_dev1, size=N) 
random_signal2 = np.random.normal(loc=mean, scale=std_dev2, size=N) 
np.random.seed(125)

plt.figure(figsize=(15, 6))
plt.plot(range(N), random_signal1, marker='o', linestyle='-',color ='red')
plt.title('  1')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()

plt.figure(figsize=(15, 6))
plt.plot(range(N), random_signal2, marker='o', linestyle='-',color ='blue')
plt.title('  2')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()
```

```python
correlation = np.correlate(random_signal1, random_signal2, mode='full')
correlation /= np.max(np.abs(correlation))  #     FFT

lags = np.arange(-N + 1, N)

plt.figure(figsize=(15, 6))
plt.plot(lags, correlation) 
plt.title(' ')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()
```

```python
#       
fft_correlation = np.fft.fft(correlation)

#       
freq = np.fft.fftfreq(len(fft_correlation))

#    
cross_spectrum_energy = np.abs(fft_correlation) ** 2

plt.figure(figsize=(15, 6))
plt.plot(freq, cross_spectrum_energy)
plt.title('  ')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()

```

```python
cross_spectrum = np.fft.fft(correlation)

cross_spectrum_phase = np.angle(cross_spectrum)

plt.figure(figsize=(15, 6))
plt.plot(freq, cross_spectrum_phase)
plt.title('  ')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(freq[:100], np.unwrap(cross_spectrum_phase[:100]))  # unwrap    
plt.title('  ')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()
```
