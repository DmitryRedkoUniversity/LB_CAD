```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

N = 100

mean = 0 
std_dev = 1  
random_signal = np.random.normal(loc=mean, scale=std_dev, size=N) 
# random_signal = np.random.randn(N)
np.random.seed(42)

autocorr = np.correlate(random_signal, random_signal, mode='full')
autocorr/=np.max(autocorr)

# lag = np.arange(-N + 1.5, N+0.5)  

plt.figure(figsize=(10, 5))
plt.plot(autocorr)
plt.title(' ')
plt.xlabel('')
plt.ylabel('')
plt.grid()


fft_result = np.fft.fft(autocorr)

#    
frequencies = np.fft.fftfreq(2*N)
plt.figure(figsize=(10, 5))
plt.plot(frequencies[:N], np.abs(fft_result[:N]))
plt.title('    ')
plt.xlabel(' ()')
plt.ylabel('')
# plt.ylim(-0.0001, 0.0001)
plt.grid()

```
